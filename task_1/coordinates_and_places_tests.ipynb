```python
#Places oordinates extraction
def extract_place_coords(s): #tweet string
    i = s.find("{'coordinates")
    c = s[i:s.find('}') + 1]
    c = c[c.find('[[') + 2: c.find(']]') + 1]

    coords = []
    for j in range(0,4):
        i = c.find(']') + 1
        coords.append(c[c.find('['):i])
        c = c[i:]

    for i in range(0,4):
        a = float(coords[i].replace('[','').replace(']','').split(',')[0])
        b = float(coords[i].replace('[','').replace(']','').split(',')[1])
        coords[i] = (a,b)
    return coords    
```

```python
places = tweets_df['place'].values
places_coordinates = []
for i in range(len(places)):
    print(str(i) + " ")
    e = extract_place_coords(str(places[i]))
    print(e)
    places_coordinates.append(e)
```

```python
def extract_coordinates(s):
    s = s[s.find('['):s.find(']') + 1]
    a = float(s.replace('[','').replace(']','').split(',')[0])
    b = float(s.replace('[','').replace(']','').split(',')[1])
    return (a,b)
```

```python
tweet_coordinates = []
for i in tweets_df['coordinates'].values:
    if i is not None:
        tweet_coordinates.append(extract_coordinates(str(i)))
tweet_coordinates        
```

```python
schipol_box = [4.73, 52.29, 4.77, 52.32]
amsterdam_box = [4.73, 52.29, 4.98, 52.42]
def check_coordinates(c, box):
    if(c[0] >= box[0]) and (c[0] <= box[2]):
        if(c[1] >= box[1]) and (c[1] <= box[3]):
            return True
    return False        
```

```python
count = 0
for i in tweet_coordinates:
    if check_coordinates(i, amsterdam_box) is True:
        count = count + 1
print("Tweets from schipol: %d" % count)        
```
