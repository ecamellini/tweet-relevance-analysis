```python
import pandas as pd

df=pd.read_csv("task2_data.csv")
df
```

```python
#http://pandas.pydata.org/pandas-docs/stable/visualization.html
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import mannwhitneyu
```

#\#Entities

```python
nr_entities = df['#entities']

plt.figure()
nr_entities.plot(kind='hist', logy=True)
plt.show()
```

```python
nr_entities_relevant = df[df['relevanceJudge']==1]['#entities']
nr_entities_non_relevant = df[df['relevanceJudge']==0]['#entities']
nr_entities_relevant.describe()
```

```python
nr_entities_non_relevant.describe()
```

```python
u, p_value = mannwhitneyu(nr_entities_non_relevant, nr_entities_relevant)
print("U: {} p_value: {}".format(u, p_value))
```

```python
plt.figure()
p = df.ix[:,['#entities','relevanceJudge']].groupby('relevanceJudge').mean().unstack().plot(x='relevanceJudge', y='#entities')
p.set_xlabel("relevanceJudge")
p.set_ylabel("#entities")
plt.show()
```
